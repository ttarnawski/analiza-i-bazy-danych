Korzystam z biblioteki numpy w celu stworzenia trzech wektorów oraz biblioteki matplotlib, aby narysować wykresy. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

f1 = lambda x : x**2 + 5

x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-6, 6, 100)
x3 = np.linspace(0, 5, 100)

plt.plot(x1, f1(x1))
plt.title('Wykres funkcji x^2 + 5 z przedziału -1 < x < 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.plot(x2, f1(x2))
plt.title('Wykres funkcji x^2 + 5 z przedziału -6 < x < 6')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.plot(x3, f1(x3))
plt.title('Wykres funkcji x^2 + 5 z przedziału 0 < x < 5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
